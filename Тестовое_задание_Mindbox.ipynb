```python
# Импорт библиотек
import pandas as pd
```

```python
# Функция подсчета суммы цифр любого числа
def sum_figures(numb):            # на вход подаем число
  sum = 0                         # сначалу сумму обнуляем
  while (numb != 0):              # пока не останется цифр
      sum = sum + numb % 10       # к сумме добавляе цифрку следующего разряда 
      numb = numb // 10           # запоминаем целучю часть для выделения цицрф следующего разряда
  return sum                      # возвращаем полученное значение
```

## 1 функция

Функция, которая подсчитывает число покупателей, попадающих в каждую группу, если нумерация ID сквозная и начинается с 0. На вход функция получает целое число n_customers (количество клиентов)

```python
# Функция подсчитывает число покупателей для каждой группы
# на вход функция получает целое число n_customers (количество клиентов)
def count_customers_group(n_customers):
  df = pd.DataFrame(columns=['id_group','amount'])                                # Создаем пустой датафрейм
  df = df.set_index('id_group', verify_integrity = True)                          # Делаем поле id_group индексом
  for i in range(n_customers):                                                    # Для i (ID клиента) в диапазоне от 0 до n_customers-1 
    sum = sum_figures(i)                                                          # получаем сумму цифр 
    if sum in df.index:                                                           # если строка с таким индексом существует
      df.at[sum, 'amount'] = df.at[sum, 'amount'] + 1                             # увеличиваем ее значение на 1
    else:                                                                         # иначе
      new_row = {'id_group': sum, 'amount': 1}                                    # создаем новую строку
      df = pd.concat([df, pd.DataFrame([new_row])], axis=0, ignore_index=True)    # и добавляем ее в наш датафрейм
  df = df.astype({"id_group": "Int64"})
  df = df.set_index('id_group')
  return df                                                                       # на выходе датафрейм с количеством покупателей в каждой группе
```

```python
df_con = count_customers_group(56456)    # Проверим для произвольного числа
df_con
```

## 2 функция

Функция, аналогичная первой, если ID начинается с произвольного числа. На вход функция получает целые числа: n_customers (количество клиентов) и n_first_id (первый ID в последовательности)

```python
# Функция подсчитывает число покупателей, если ID начинается с произвольного числа. 
# На вход функция получает целые числа: n_customers (количество клиентов) и n_first_id (первый ID в последовательности)
def count_customers_group1(n_customers,  n_first_id):
  df =  pd.DataFrame(columns=['id_group','amount'])
  df = df.set_index('id_group', verify_integrity = True)
  for i in range(n_first_id, n_first_id + n_customers):                           # Для i (ID клиента) в диапазоне от n_first_id до полного количества клиентов
    sum = sum_figures(i)
    if sum in df.index:
      df.at[sum, 'amount'] = df.at[sum, 'amount'] + 1
    else:
      new_row = {'id_group': sum, 'amount': 1} 
      df = pd.concat([df, pd.DataFrame([new_row])], axis=0, ignore_index=True)
  df = df.astype({"id_group": "Int64"})
  df = df.set_index('id_group')
  return df

# По идее и эту функцию можно использовать вместо первой при n_first_id = 0, 
# можно задать его дефлолтным
```
